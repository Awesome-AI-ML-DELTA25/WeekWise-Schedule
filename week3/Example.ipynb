```python
import matplotlib.pyplot as plt
import numpy as np
```

### And Function

Perceptrons good at classifying linearly seperable data!

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]])
y = np.array([0, 0, 0, 1])
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], color='red', label='Class 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Class 1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('AND Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def sign(x):
    return 1 if x > 0 else 0

def f(x):
    w = np.array([4/5, 4/5])
    b = -1
    return sign(np.dot(w, x) + b)
```

```python
# Plot decision boundary
x1 = np.linspace(-0.5, 1.5, 100)
x2 = np.linspace(-0.5, 1.5, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = np.array([f(np.array([x1, x2])) for x1, x2 in zip(X1.ravel(), X2.ravel())])
Z = Z.reshape(X1.shape)
plt.contourf(X1, X2, Z, alpha=0.5, levels=[-0.1, 0.5, 1.1], colors=['red', 'blue'])
plt.scatter(X[y == 0, 0], X[y == 0, 1], color='red', label='Class 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Class 1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Decision Boundary for AND Function')
plt.legend()
plt.grid(True)
plt.savefig('./images/AND.png')
plt.show()
```

What about XOR Function?

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]])
y = np.array([0, 1, 1, 0])
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], color='red', label='Class 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Class 1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('AND Function')
plt.legend()
plt.grid(True)
plt.show()
```

Solution: Use two layers !!

```python
def sign(x):
    return 1 if x > 0 else 0

def AND(x):
    w = np.array([4/5, 4/5])
    b = -1
    return sign(np.dot(w, x) + b)

def OR(x):
    w = np.array([2, 2])
    b = -1
    return sign(np.dot(w, x) + b)

def XOR(x):
    w = np.array([-1, 1])
    b = 0
    x_ = np.array([AND(x), OR(x)])
    return sign(np.dot(w, x_) + b)
```

```python
# Plot decision boundary
x1 = np.linspace(-0.5, 1.5, 100)
x2 = np.linspace(-0.5, 1.5, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = np.array([XOR(np.array([x1, x2])) for x1, x2 in zip(X1.ravel(), X2.ravel())])
Z = Z.reshape(X1.shape)
plt.contourf(X1, X2, Z, alpha=0.5, levels=[-0.1, 0.5, 1.1], colors=['red', 'blue'])
plt.scatter(X[y == 0, 0], X[y == 0, 1], color='red', label='Class 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Class 1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Decision Boundary for XOR Function')
plt.legend()
plt.grid(True)
plt.savefig('./images/XOR.png')
plt.show()
```

```python
def ReLU(x):
    return np.maximum(0, x)
def XOR2(x):
    W1 = np.array([[1, 1], [1, 1]])
    b1 = np.array([0, -1])
    W2 = np.array([1, -2])
    b2 = -0.5

    hidden = ReLU(np.dot(W1, x) + b1)
    output = np.dot(W2, hidden) + b2

    return sign(output)
```

```python
# Plot decision boundary
x1 = np.linspace(-0.5, 1.5, 100)
x2 = np.linspace(-0.5, 1.5, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = np.array([XOR2(np.array([x1, x2])) for x1, x2 in zip(X1.ravel(), X2.ravel())])
Z = Z.reshape(X1.shape)
plt.contourf(X1, X2, Z, alpha=0.5, levels=[-0.1, 0.5, 1.1], colors=['red', 'blue'])
plt.scatter(X[y == 0, 0], X[y == 0, 1], color='red', label='Class 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Class 1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Decision Boundary for XOR2 Function')
plt.legend()
plt.grid(True)
plt.savefig('./images/XOR2.png')
plt.show()
```
