```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 2*x + 1
```

```python
x = np.linspace(0, 5, 50)
y = f(x) + np.random.normal(0, 0.2, size=x.shape)
plt.scatter(x, y, label='Real world data')
plt.plot(x, f(x), color='red', label='Good Approximation')
plt.title('Scatter plot with a line of best fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Why AI/ML?
(Supervised) AI/ML looks for patterns in the data and uses them to make predictions. In the graph shown above, you can see that the above graph looks approximately like a linear function. You were able to exploit the visualisation on a 2-D graph to see that the data is linear, despite the fact that the raw data didn't make any sense. A sharp eye may infact be able to see that f(x) = 2x+1 is a good approximation of the data. The data is not linear, but it is close enough to be approximated by a linear function. This is the key to AI/ML.

Such a thing is hard to visualise in 3-D or higher dimensions. The data maybe linear, but it's not trivial to capture the data in a linear function. The AI/ML algorithm is able to do this for you. All of this depends on how data is represented. 

AI/ML gives you a good decision rule for a given set of data. Based on the data, it will give you a function that is a good approximation of the data.

Some applications:
- Predicting the price of a house based on its features (e.g., number of bedrooms, square footage, etc.)
- Predicting the price of a stock based on its historical prices
- Predicting the outcome of a sports game based on the teams' statistics
- Answering whether a given picture contains a cat or not
